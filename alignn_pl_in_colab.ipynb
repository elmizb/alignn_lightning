```python
import json
import torch
import pandas as pd
from sklearn.metrics import mean_squared_error
import pytorch_lightning as pl
from pytorch_lightning.callbacks.early_stopping import EarlyStopping
# from pytorch_lightning.callbacks import StochasticWeightAveraging
# from pytorch_lightning.tuner import Tuner
from alignn_pl.data_pl import ALIGNNDataModule
from alignn_pl.config import ALIGNNConfig as config
# from alignn_pyg.model_pl import ALIGNN # ALIGNN model with pytorch_geometric
from alignn_pl.alignn import ALIGNN # model with dgl
```

### download the jarvis datasets (jarvis_dft3d_2021)  
In the original paper of ALIGNN, the version of jarvis_dft datasets is 2021.  
There is useful method in jarvis-tools to download the datasets but it can't do it for "jarvis_dft3d_2021"

```python
!mkdir data
!wget https://figshare.com/ndownloader/files/26808914 -O ./data/jarvis_dft3d_2021.zip
```

```python
!unzip ./data/jarvis_dft3d_2021.zip -d ./data/
```

```python
with open("./data/d3-3-12-2021.json", 'r') as f:
    data_json=json.load(f)
ds = pd.DataFrame(data_json)
```

```python
ds.head(2)
```

keep only the necessary columns(id, structure data, target label)  
and divide 8:1:1 for training, validation and test.

```python
target = "formation_energy_peratom"
ds = ds[["jid", "atoms", target]]

n_train = int(0.8*len(ds))
n_valid = int(0.1*len(ds))
train = ds[:n_train]
valid = ds[n_train:n_train+n_valid]
test = ds[n_valid+n_train:]
```

save them

```python
train.to_pickle("./data/train_dft3d.pkl")
valid.to_pickle("./data/valid_dft3d.pkl")
test.to_pickle("./data/test_dft3d.pkl")
```

### crystal structure text (atoms, structure) -> graph, line graph  
It takes long time (pandas is in...) I recommend to do it separately (on local computer).  
In that case, after generate .data file which is including graphs,   
put them in ./data and tell the name(path) without the tail part "train(/valid/test/).data" in data_dir of ALIGNNDataModule. 

```python
jarvis_data = ALIGNNDataModule(
    data_dir="./data/jarvis_dgl_",
    graph_fmt="dgl"
)
```

```python
jarvis_data.prepare_data(
    df_train=train,
    df_valid=valid,
    df_test=test,
    use_pandarallel=True
)
```

setup the data for training

```python
jarvis_data.setup(
    stage="fit", 
    minmax_scale=False,
    standard_scale=True
    )
```

```python
trainloader = jarvis_data.train_dataloader()
valloader = jarvis_data.val_dataloader()
```

### Training with pytorch_lightning

```python
trainer = pl.Trainer(max_epochs=100,
                    # precision=16,
 
                    #profiler="pytorch",
                    callbacks=[
                        EarlyStopping(monitor="val_loss", mode="min", verbose=True, patience=10),
                        # StochasticWeightAveraging(swa_lrs=1e-2)
                        ]
                    )
```

### fitting

```python
trainer.fit(
    model=ALIGNN(),
    train_dataloaders=trainloader,
    val_dataloaders=valloader

)
```

save trainig results

```python
trainer.save_checkpoint("./trainer.chkpt")
```

load trainer

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
check_point = torch.load("/content/alignn_lightning/trainer.chkpt" ,map_location=torch.device(device))
```

```python
model_weights = check_point["state_dict"]
```

```python
model = ALIGNN()
model.load_state_dict(model_weights)
```

```python
jarvis_data.setup(
    stage="test",
    )
test_dataloader = jarvis_data.test_dataloader()
```

inference for test data

```python
trainer = pl.Trainer()
predictions = trainer.predict(model, test_dataloader)
```

The scaler coefficient is made at jarvis_data.setup at "fit" stage.  
If you started with load checkpoint, you need to execute setup with "fit".  

```python
label_mean = torch.load("./data/jarvis_dgl_label_mean.data")
label_std = torch.load("./data/jarvis_dgl_label_std.data")

corrected_predictions = [pred * label_std + label_mean for pred in predictions]
```

```python
mean_squared_error(jarvis_data.dataset_test.labels.numpy(), corrected_predictions, squared=False)
```
